```python
from IPython.display import display

Blue=len(data[(data['HOURLY REQ']>=4) ])
Green=len(data[(data['HOURLY REQ']<4)& (data['HOURLY REQ']>=3) ])
Yellow=len(data[(data['HOURLY REQ']<3 ) & (data['HOURLY REQ']>=2) ])
Red=len(data[(data['HOURLY REQ']<2) ])

TM=(f"{((data['MODULE#SECTION'].count()/data['MODULE#SECTION'].count()*100).astype(int))}%")
Bluep=(f"{((Blue/data['MODULE#SECTION'].count()*100).round(1))}%")
Greenp=(f"{((Green/data['MODULE#SECTION'].count()*100).round(1))}%")
Yellowp=(f"{((Yellow/data['MODULE#SECTION'].count()*100).round(1))}%")
Redp=(f"{((Red/data['MODULE#SECTION'].count()*100).round(1))}%")



# Define the DataFrame
data4 = pd.DataFrame({
    'Information': ['Total Modules', '', '', '', ''],
    'No of Modules': [data['MODULE#SECTION'].count(), Blue, Green, Yellow, Red],  # Just sample data, replace with your actual data
    'Percentage': [ TM,Bluep, Greenp, Yellowp, Redp],
}, index=pd.Index([1, 2, 3, 4, 5]))

# Function to apply background color conditionally
def color_specific_cell_blue(data):
    # Create a new DataFrame that's the same size as the input, filled with blank styles
    styles = pd.DataFrame('', index=data.index, columns=data.columns)
    # Set the background color of the specific cell to blue
    styles.loc[2, 'Information'] = 'background-color: #00C0FF'
    styles.loc[3, 'Information'] = 'background-color: #50D092'
    styles.loc[4, 'Information'] = 'background-color: #F0B000'
    styles.loc[5, 'Information'] = 'background-color: Red' # 2 here is the index and 'No of Modules' is the column name
    return styles

# Apply the style function to the DataFrame
styled_data4 = data4.style.apply(color_specific_cell_blue, axis=None)


# Display the styled DataFrame
display(styled_data4)


# Normalize the 'HOURLY REQ' values to have equal width bars
max_hourly_req = data['HOURLY REQ'].max()
data['Normalized'] = max_hourly_req

# Create the plot
plt.figure(figsize=(10, 8))
ax = data.plot(kind='barh', x='MODULE#SECTION', y='Normalized', legend=False, color=data["Status"], fontsize=14)
ax.set_xlabel("WIP", fontsize=14)  # Set font size for the x-axis label
ax.set_title("Horizontal Bar Chart with Multiple Labels per Bar", fontsize=16)  # Set font size for the title

# Adding custom labels inside the bars
for idx, rect in enumerate(ax.patches):
    # Calculate the center position of the bar
    y = rect.get_y() + rect.get_height() / 2
    half_height = rect.get_height() / 4  # Divide bar height into segments for each label

    # Display Module Section and HOURLY REQ inside the bar
    label_module = f'{data.iloc[idx]["MODULE#SECTION"]} - {data.iloc[idx]["HOURLY REQ"]} WIP'
    ax.text(rect.get_width() / 2, y, label_module, ha='center', va='center', fontsize=14, color='Black')

    # For DAILY REQ and Issued to Module (outside the bar)
    label_daily_req = f'DR: {data.iloc[idx]["DALLY REQ"]}'
    label_issued = f'HR: {data.iloc[idx]["Issued to Module"]}'
    ax.text(rect.get_width() + 10, y + half_height, label_daily_req, ha='left', va='center', fontsize=14)
    ax.text(rect.get_width() + 10, y - half_height, label_issued, ha='left', va='center', fontsize=14)

# Adjust the x-axis limit to make space for text
ax.set_xlim(0, max_hourly_req * 1.5)  # Increase x-axis limit for text

# Show the plot
plt.show()

```
